```python
#categorizing vote_avergae column
#we would cut the vote_average values and make 4 categories:popular avearage below_avg not_popular to describe it more using catigorize_col() function provided above
```

```python
def catigorize_col(df,col,labels):
  edges=[df[col].describe()['min'],
         df[col].describe()['25%'],
         df[col].describe()['50%'],
         df[col].describe()['75%'],
         df[col].describe()['max']]
  df[col]= pd.cut(df[col],edges,labels=labels,duplicates='drop')
  return df
```

```python
labels=['not_popular','below_avg','average','popular']
catigorize_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```
